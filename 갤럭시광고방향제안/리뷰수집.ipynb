```python
import re
import time
import pandas as pd
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.common.keys import Keys 


# 브라우저 열기
browser = webdriver.Chrome()

# 네이버 쇼핑 페이지 열기
url = 'https://search.shopping.naver.com/catalog/38117680618?query=%EC%82%BC%EC%84%B1%20%EA%B0%A4%EB%9F%AD%EC%8B%9C%20s23&NaPm=ct%3Dlp1dl3kg%7Cci%3D5dc892fb1627c9891100e1042d37801ac6bbdaa8%7Ctr%3Dslsl%7Csn%3D95694%7Chk%3D3c3f923599b2bc141b002524d44b7be2bd637340'
browser.get(url)

# 결과 저장 리스트 초기화
results = []

# 페이지 이동 및 리뷰 수집
page = 1  # 페이지 변수 초기화
while page <= 50:  # 페이지 변수가 50 이하일 때까지 반복
    page_btn_list = browser.find_elements('css selector', 'div.review_section_review__GDdvh > div.pagination_pagination__JW7zT > a')
    for page_btn in page_btn_list:
        if page_btn.text == "다음":
            page_btn.click()  # 다음 페이지로 이동
            time.sleep(1)
            break  # 다음 페이지로 이동했으면 내부 반복문 종료

        page_number_match = re.search(r'\d+', page_btn.text)  # 페이지 번호에서 숫자 추출
        if page_number_match is not None:
            page_number = page_number_match.group()
            if int(page_number) == page:  # 페이지 변수와 일치하는 경우에만 클릭
                print(page_btn.text)
                page_btn.click()
                time.sleep(1)

                html = browser.page_source
                soup = BeautifulSoup(html, 'html.parser')
                
                # 리뷰 리스트 선택자 확인
                review_list = soup.select('ul.reviewItems_list_review__q726A > li')
                
                # 리뷰 정보 수집 부분 확인
                for review in review_list:
                    star = review.select('span.reviewItems_average__0kLWX')[0].text
                    date = review.select('span.reviewItems_etc__9ej69')[2].text
                    try:
                        info = review.select('span.reviewItems_etc__9ej69')[3].text
                    except:
                        info = '정보없음'  # 색상, 모델 기입 안 되어 있는 경우 있음.
                    content = review.select('p.reviewItems_text__XrSSf')[0].text

                    data = [star, date, info, content]
                    results.append(data)

                break  # 현재 페이지에서 정보를 수집했으면 내부 반복문 종료

    page += 1  # 페이지 변수 증가

# 데이터프레임 형태로 변환
df = pd.DataFrame(results, columns=['star', 'date', 'info', 'content'])

# df 내용 csv 파일로 저장
df.to_csv('./갤럭시23리뷰수집.csv', index=False)

# 브라우저 종료
browser.quit()
```
